# Advanced Lane Finding Project

#### The goals / steps of this project are the following:

Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
Apply a distortion correction to raw images.
Use color transforms, gradients, etc., to create a thresholded binary image.
Apply a perspective transform to rectify binary image ("birds-eye view").
Detect lane pixels and fit to find the lane boundary.
Determine the curvature of the lane and vehicle position with respect to center.
Warp the detected lane boundaries back onto the original image.
Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

```python
def apply_filters(img):
    sobel_kernel=3
    angle_threshold =(0.8, 1.2)
    magnitude_threshold =(45,255) 
    sat_threshold =(75,255) 
    lightness_threshold = 40 
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    absx = np.abs(sobel_x)
    absy = np.abs(sobel_y)
    
    angles = np.arctan2(absy, absx)
    magnitude = np.sqrt(sobel_x**2+sobel_y**2)
    scaled_magnitude = np.uint8(255*magnitude/np.max(magnitude))

    hls = cv2.cvtColor( img, cv2.COLOR_BGR2HLS )
    L = hls[:,:,1]
    S = hls[:,:,2]

    binary_output = np.zeros_like(scaled_mag)
    binary_output[(L > lightness_threshold) & 
                  (((scaled_mag >= magnitude_threshold[0]) &
                     (scaled_mag <= magnitude_threshold[1]) &   
                     (angles >= angle_threshold[0]) &   
                     (angles <= angle_threshold[1])) |
                     ((S > sat_threshold[0]) &
                     (S <= sat_threshold[1])))] = 1

    return binary_output

test_image_BGR_1 = cv2.imread('test_images/test6.jpg')
plt.imshow(apply_filters(test_image_BGR_1), cmap='gray')
```

```python
def transform_perspective(img):
    h, w, _ = img.shape
    
    src = np.float32([(575,464),
                  (707,464), 
                  (258,682), 
                  (1049,682)])
    dst = np.float32([(450,0),
                      (w-450,0),
                      (450,h),
                      (w-450,h)])
    
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    return cv2.warpPerspective(img, M, (w,h), flags=cv2.INTER_LINEAR), Minv

plt.imshow(transform_perspective(plt.imread('test_images/straight_lines1.jpg'))[0])
```

```python
def pipeline(img):
    undistorted = cv2.undistort(img, camera_matrix, dist, None, camera_matrix)
    tfed, Minv = transform_perspective(undistorted)
    return apply_filters(tfed), Minv
    
raw = plt.imread('test_images/straight_lines1.jpg')

plt.imshow(pipeline(raw)[0], cmap='gray')
```

```python
def sliding_windows(binary_warped, show_image=False):
    
    histogram = np.sum(binary_warped[int(binary_warped.shape[0]/2):,:], axis=0)
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    midpoint = np.int(histogram.shape[0]/2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    nwindows = 9
    window_height = np.int(binary_warped.shape[0]/nwindows)
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    leftx_current = leftx_base
    rightx_current = rightx_base
    margin = 40
    minpix = 25
    left_lane_inds = []
    right_lane_inds = []
    
    for window in range(nwindows):
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)).nonzero()[0]
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
            
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds]
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds] 
    
    left_fit, right_fit = None, None
    if len(leftx) > 0:
        left_fit = np.polyfit(lefty, leftx, 2)
    if len(rightx) > 0:
        right_fit = np.polyfit(righty, rightx, 2)
    
    if show_image:
        ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0])
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
        plt.imshow(out_img)
        plt.plot(left_fitx, ploty, color='yellow')
        plt.plot(right_fitx, ploty, color='yellow')
    
    return left_fit, right_fit, leftx, lefty, rightx, righty
    
    
```

```python
def polyfit_with_prev_fits(binary_warped, left_prev, right_prev, show_image=False):
    
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    margin = 30
    left_lane_inds = ((nonzerox > (left_prev[0]*(nonzeroy**2) + left_prev[1]*nonzeroy + left_prev[2] - margin)) & 
                      (nonzerox < (left_prev[0]*(nonzeroy**2) + left_prev[1]*nonzeroy + left_prev[2] + margin))) 
    right_lane_inds = ((nonzerox > (right_prev[0]*(nonzeroy**2) + right_prev[1]*nonzeroy + right_prev[2] - margin)) & 
                       (nonzerox < (right_prev[0]*(nonzeroy**2) + right_prev[1]*nonzeroy + right_prev[2] + margin)))  

    # Again, extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    left_fit_new, right_fit_new = None, None
    if len(leftx) > 0:
        left_fit_new = np.polyfit(lefty, leftx, 2)
    if len(rightx) > 0:
        right_fit_new = np.polyfit(righty, rightx, 2)

    if show_image:
        ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0])
        left_fitx = left_fit_new[0]*ploty**2 + left_fit_new[1]*ploty + left_fit_new[2]
        right_fitx = right_fit_new[0]*ploty**2 + right_fit_new[1]*ploty + right_fit_new[2]
        # Create an image to draw on and an image to show the selection window
        out_img = np.dstack((binary_warped, binary_warped, binary_warped))*255
        window_img = np.zeros_like(out_img)
        # Color in left and right line pixels
        out_img[nonzeroy[left_lane_inds], nonzerox[left_lane_inds]] = [255, 0, 0]
        out_img[nonzeroy[right_lane_inds], nonzerox[right_lane_inds]] = [0, 0, 255]

        left_line_window1 = np.array([np.transpose(np.vstack([left_fitx-margin, ploty]))])
        left_line_window2 = np.array([np.flipud(np.transpose(np.vstack([left_fitx+margin, 
                                      ploty])))])
        left_line_pts = np.hstack((left_line_window1, left_line_window2))
        right_line_window1 = np.array([np.transpose(np.vstack([right_fitx-margin, ploty]))])
        right_line_window2 = np.array([np.flipud(np.transpose(np.vstack([right_fitx+margin, 
                                      ploty])))])
        right_line_pts = np.hstack((right_line_window1, right_line_window2))

        cv2.fillPoly(window_img, np.int_([left_line_pts]), (0,255, 0))
        cv2.fillPoly(window_img, np.int_([right_line_pts]), (0,255, 0))
        result = cv2.addWeighted(out_img, 1, window_img, 0.3, 0)
        plt.imshow(result)
        plt.plot(left_fitx, ploty, color='yellow')
        plt.plot(right_fitx, ploty, color='yellow')
        plt.xlim(0, 1280)
        plt.ylim(720, 0)
    
    return left_fit_new, right_fit_new, leftx, lefty, rightx, righty
```

```python
example_img = cv2.imread('./test_images/test3.jpg')
example_img = cv2.cvtColor(example_img, cv2.COLOR_BGR2RGB)
example_img_bin, Minv = pipeline(example_img)
    
left_fit, right_fit, *_ = sliding_windows(example_img_bin, True)
_, _, *points = polyfit_with_prev_fits(example_img_bin, left_fit, right_fit)
```

```python
# measure curvature ... see 
# https://classroom.udacity.com/nanodegrees/nd013/parts/fbf77062-5703-404e-b60c-95b78b2f3f9e/modules/2b62a1c3-e151-4a0e-b6b6-e424fa46ceab/lessons/096009a1-3d76-4290-92f3-055961019d5e/concepts/2f928913-21f6-4611-9055-01744acc344f

def calc_curv_rad(bin_img, leftx, lefty, rightx, righty):
    h, w = bin_img.shape
    ym_per_pix = 3.048/100 # meters per pixel in y dimension
    xm_per_pix = 3.7/370 # meters per pixel in x dimension
    
    if len(leftx) <= 0 or len(rightx) <= 0:
        return None, None, None

    y_eval = h
    # Fit new polynomials to x,y in world space
    left_fit_cr = np.polyfit(lefty*ym_per_pix, leftx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(righty*ym_per_pix, rightx*xm_per_pix, 2)
    # Calculate the new radii of curvature
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    
    h_scaled = h*ym_per_pix
    l_fit_start = left_fit_cr[0]*h_scaled**2 + left_fit_cr[1]*h_scaled + left_fit_cr[2]
    r_fit_start = right_fit_cr[0]*h_scaled**2 + right_fit_cr[1]*h_scaled + right_fit_cr[2]
    lane_center = (l_fit_start + r_fit_start)/2
    car_center = w/2*xm_per_pix
    dist_from_center = w/2*xm_per_pix - lane_center
    return left_curverad, right_curverad, dist_from_center
```

```python
radl, radr, dist_from_center = calc_curv_rad(example_img_bin, *points)
```

```python
# draw the overlay on the original image
# also add text

def draw_lane(original_img, binary_img, l_fit, r_fit, Minv):
    new_img = np.copy(original_img)
    if l_fit is None or r_fit is None:
        return original_img
    
    warp_zero = np.zeros_like(binary_img, dtype=np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    h,w = binary_img.shape
    ploty = np.linspace(0, h-1, num=h)
    left_fitx = l_fit[0]*ploty**2 + l_fit[1]*ploty + l_fit[2]
    right_fitx = r_fit[0]*ploty**2 + r_fit[1]*ploty + r_fit[2]

    pts_left = np.array([np.vstack([left_fitx, ploty]).T], dtype=np.int32)
    pts_right = np.array([np.flipud(np.vstack([right_fitx, ploty]).T)], dtype=np.int32)
    pts = np.array(np.hstack((pts_left, pts_right)).astype(np.int_))

    cv2.fillPoly(color_warp, pts, (0, 255, 0))
    cv2.polylines(color_warp, pts_left, isClosed=False, color=(255,0,255), thickness=10)
    cv2.polylines(color_warp, pts_right, isClosed=False, color=(0,255,255), thickness=10)

    unwarped = cv2.warpPerspective(color_warp, Minv, (w, h)) 
    return cv2.addWeighted(new_img, 1, unwarped, 0.5, 0)

def draw_text(original_img, radl, radr, dist_from_center):
    if not radl or not radr or not dist_from_center:
        return original_img
    
    curv_rad = (radl+radr)/2
    
    scale = 2
    thickness = 2
    color = (200, 255, 155)
    new_img = np.copy(original_img)
    h = new_img.shape[0]
    font = cv2.FONT_HERSHEY_PLAIN
    text = 'Curve radius: ' + '{:04.2f}'.format(curv_rad) + 'm'
    cv2.putText(new_img, text, (40,40), font, scale, color, thickness, cv2.LINE_AA)
    
    direction = ''
    if dist_from_center > 0:
        direction = 'right'
    elif dist_from_center < 0:
        direction = 'left'
    text = '{:04.3f}'.format(abs(dist_from_center)) + 'm ' + direction + ' of center'
    cv2.putText(new_img, text, (40,80), font, scale, color, thickness, cv2.LINE_AA)
    return new_img
```

```python
old_points = None
old_left = None
old_right = None

def points_too_different(points1, points2):
    if (len(points1[0]) < 1  or len(points2[0]) < 1):
        print("dont have enough points")
        return False
    a1, a2, a3, a4 = points1
    b1, b2, b3, b4 = points2
    return abs(np.max(a3)-np.max(b3)) > 50 \
            or abs(np.min(a3)-np.min(b3)) > 50
        
def has_no_points(points):
    for x in points:
        if len(x) == 0:
            return True
    return False


def generate_image(img):
    global old_points 
    global old_left
    global old_right
    
    new_img = np.copy(img)
    img_bin, Minv = pipeline(new_img)
    
    left_fit, right_fit, *points = sliding_windows(img_bin)
    
    if has_no_points(points):
        print("no points")
        sliding_windows(img_bin, True)
        points = old_points
        left_fit = old_left
        right_fit = old_right
    
    if (old_points and points_too_different(points, old_points)):
        print("bad points")
        left_fit, right_fit, *points = polyfit_with_prev_fits(img_bin, old_left, old_right)
        
    radl, radr, dist_from_center = calc_curv_rad(img_bin, *points)
    with_lane = draw_lane(new_img, img_bin, left_fit, right_fit, Minv)
    with_text = draw_text(with_lane, radl, radr, dist_from_center)
    
    old_points = points
    old_left = left_fit
    old_right = right_fit
        
    return with_text
        
```
